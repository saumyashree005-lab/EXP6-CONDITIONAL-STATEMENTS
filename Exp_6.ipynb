```python
#5. To find the given year is a leap year or not
year = int(input("Enter a year:"))
if (year %4 == 0 and year%100!=0) or year%400 == 0:
  print(year,"is a leap year")
else:
  print(year,"is not a leap year")
```

```python
#6. Write a python program to make increement in the date
# Input date from user
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

# Number of days in each month
days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# Check leap year
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    days[1] = 29   # February has 29 days in leap year

# Increase day by 1
day = day + 1

# Check if day exceeds days in month
if day > days[month - 1]:
    day = 1
    month = month + 1

    # Check if month exceeds December
    if month > 12:
        month = 1
        year = year + 1

# Print next date
print("Next Date:", day, "-", month, "-", year)


```
