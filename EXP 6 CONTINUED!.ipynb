```python
#PROBLEM STATEMENT!
#write a python program to calculate the gross salary of an employee based
#if basic = 10000: HRA = 20%, DA = 80%
#if basic = 20000: HRA = 25%, DA = 20%
#otherwise: HRA = 30%, DA = 95%
basic = float(input("Enter basic salary: "))
if basic == 10000:
    hra = basic * 0.20
    da = basic * 0.80
elif basic == 20000:
    hra = basic * 0.25
    da = basic * 0.20
else:
    hra = basic * 0.30
    da = basic * 0.95
gross_salary = basic + hra + da
print("Basic Salary:", basic)
print("Gross Salary:", gross_salary)
```

```python
#Write a python program to calculate income tax based on annual income
#up to rupees 2,50,000 : no tax
#rupees 2,50,000-5,00,000 : 5%
#5,00,000-10,00,000 : 20%
#above 10,00,000 : 30%
income = float(input("Enter annual income (in rupees): "))
tax = 0
if income <= 250000:
    tax = 0
elif income <= 500000:
    tax = (income - 250000) * 0.05
elif income <= 1000000:
    tax = (250000 * 0.05) + (income - 500000) * 0.20
else:
    tax = (250000 * 0.05) + (500000 * 0.20) + (income - 1000000) * 0.30

print("Income Tax to be paid: ₹", tax)








```
